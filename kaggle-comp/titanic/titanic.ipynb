```python
# Importing packages
import numpy as np
import pandas as pd
```

```python
# Reading training data
train_df = pd.read_csv("./train.csv")
train_df.head()
```

```python
# Strategy 1, finding correlations to base predictions on
corr_df = train_df

# Removing nonsense columns
corr_df.drop(columns=['Name', 'Ticket', 'Cabin', "Embarked"], inplace=True)

# Replacing features
corr_df.loc[corr_df['Sex'] == 'female', 'Sex'] = 1
corr_df.loc[corr_df['Sex'] == 'male', 'Sex'] = 0
corr_df.corr()
```
